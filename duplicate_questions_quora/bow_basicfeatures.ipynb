### ✍️ Features for Question Pair Similarity

1. **q1 len** → character length of question 1
2. **q2 len** → character length of question 2
3. **q1 words** → number of words in question 1
4. **q2 words** → number of words in question 2
5. **words common** → number of common unique words
6. **words total** → total number of words in q1 + total number of words in q2
7. **word share** → `words common / words total`

```python
# Feature Engineering

new_df['q1_len'] = new_df['question1'].str.len()
new_df['q2_len'] = new_df['question2'].str.len()
```

```python
new_df['q1_num_words'] = new_df['question1'].apply(lambda row: len(row.split(" ")))
new_df['q2_num_words'] = new_df['question2'].fillna('').apply(lambda row: len(row.split(" ")))
new_df.head()
```

```python
def common_words(row):
    w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), str(row['question2']).split(" ")))
    return len(w1 & w2)
```

```python
new_df['word_common'] = new_df.apply(common_words, axis=1)
new_df.head()
```

```python
def total_words(row):
    q1 = str(row['question1']) if pd.notnull(row['question1']) else ""
    q2 = str(row['question2']) if pd.notnull(row['question2']) else ""

    w1 = set(map(lambda word: word.lower().strip(), q1.split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), q2.split(" ")))

    return len(w1) + len(w2)
```

```python
new_df['word_total'] = new_df.apply(total_words, axis=1)
new_df.head()
```

```python
new_df['word_share'] = round(new_df['word_common']/new_df['word_total'],2)
new_df.head()
```
